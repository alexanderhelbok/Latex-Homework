```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 2.5, 5, 10, 15, 20])
y = np.array([0, 59, 200, 725, 1568, 2720])	

def func(x, a, b):
    return a*x**2 + b*x

def para(x, a, b, c):
    return a*x**2 + b*x + c

def linear(x, a, b):
    return a*x + b

popt, pcov = curve_fit(func, x, y)
print(popt)

a, b = -1/0.015, 1177

plt.plot(x, y, 'ko', label="Original Data")
plt.plot(x, func(x, *popt), 'r-', label="Fitted Curve")
plt.plot(x, -linear(x, a, b), 'b-', label="Linear Curve")
# add both functions to the plot
plt.plot(x, -linear(x, a, b) + func(x, *popt), 'b-', label="Linear Curve")
plt.plot(x, para(x, popt[0], popt[1] - a, -b))

print(-linear(x, a, b) + func(x, *popt))
```
